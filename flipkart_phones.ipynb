#### DATA GATHERING

We will gather our data using **Selenium** to automate the browser and load dynamic web content and then use **BeautifulSoup** to parse the HTML and extract the desired information.

1. Importing NumPy library for numerical operations.
2. Importing pandas library for data manipulation and analysis.
3. Importing time module for time-related functions.
4. Importing the WebDriver class from the Selenium library for browser automation.
5. Importing the Service class to manage the ChromeDriver executable.
6. Importing the By class to locate elements on a web page.
7. Importing BeautifulSoup from the bs4 library for parsing HTML and XML documents.

```python
import numpy as np
import pandas as pd
import time
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
```

The following code initializes a Selenium WebDriver session for Chrome by specifying the path to the chromedriver.exe using a Service object, allowing automated browser interactions.

```python
s = Service('C:/Users/Shivani/Desktop/chromedriver.exe')
driver = webdriver.Chrome(service = s)
```

Creating dataframe named 'samsung' for Samsung mobiles

```python
df_samsung = pd.DataFrame()

for j in range(1,32):
    driver.get('https://www.flipkart.com/search?q=samsung+mobiles&otracker=search&otracker3=search&marketplace=FLIPKART&as-show=on&as=off&page={}'.format(j))
    time.sleep(1)
    html = driver.page_source
    soup = BeautifulSoup(html, 'lxml')

    mobile_samsung = soup.find_all('div',class_ = 'tUxRFH')

    image = []
    name = []
    rating = []
    rating_review = []
    ram_rom = []
    display = []
    camera = []
    battery = []
    processor = []
    warranty = []
    original_price = []
    discount = []
    discounted_price = []
    
    for i in mobile_samsung:
        try:
            image.append(i.find('img', class_="DByuf4")['src'])
        except:
            image.append(np.nan)
        try:
            name.append(i.find('div', class_ ='KzDlHZ').text.strip())
        except:
            name.append(np.nan)
        try:
            rating.append(i.find('div', class_ = 'XQDdHH').text.strip())
        except:
            rating.append(np.nan)
        try:
            rating_review.append(i.find('span', class_='Wphh3N' ).text.strip())
        except:
            rating_review.append(np.nan)
        try:
            ram_rom.append(i.find_all('li', class_='J+igdf' )[0].text.strip())
        except:
            ram_rom.append(np.nan)
        try:
            display.append(i.find_all('li', class_='J+igdf' )[1].text.strip())
        except:
            display.append(np.nan)
        try:
            camera.append(i.find_all('li', class_='J+igdf')[2].text.strip())
        except:
            camera.append(np.nan)
        try:
            battery.append(i.find_all('li', class_='J+igdf')[3].text.strip())
        except:
            battery.append(np.nan)
        try:
            processor.append(i.find_all('li', class_='J+igdf')[4].text.strip())
        except:
            processor.append(np.nan)
        try:
            original_price.append(i.find('div', class_='yRaY8j ZYYwLA').text.strip())
        except:
            original_price.append(np.nan)
        try:
            discount.append(i.find('div', class_='UkUFwK').text.strip())
        except:
            discount.append(np.nan)
        try:
            warranty.append(i.find_all('li', class_='J+igdf')[5].text.strip())
        except:
            warranty.append(np.nan)
        try:
            discounted_price.append(i.find('div', class_='Nx9bqj _4b5DiR').text.strip())
        except:
            discounted_price.append(np.nan)

    
    d = {
        'image' : image,
        'name' : name,
        'rating' : rating,
        'rating_review' : rating_review,
        'ram_rom' : ram_rom,
        'display' : display,
        'camera' : camera,
        'battery' : battery,
        'processor' : processor,
        'warranty' : warranty,
        'original_price' : original_price,
        'discount' : discount,
        'discounted_price' : discounted_price
    }
    
    df = pd.DataFrame(d)
    
    df_samsung = pd.concat([df_samsung, df], ignore_index=True)

df_samsung
```

Converting our dataframe into a CSV file so that we can import it later for cleaning.

```python
df_samsung.to_csv('C:/Users/Shivani/Downloads/flips.csv', index= False)
```
